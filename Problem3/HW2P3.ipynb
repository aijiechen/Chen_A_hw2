## Problem 3: Is the average of polls better than just one poll?

#### Problem 3(a)

Most undecided voters vote for one of the two candidates at the election. Therefore, the reported percentages underestimate the final value of both candidates. However, if we assume the undecided will split evenly, then the observed difference should be an unbiased estimate of the final difference. 

Add a new column to the `election` DataFrame containg the difference between Obama and Romeny called `Diff`. 

```python
#your code here
load= requests.get('http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv')
election=pd.read_csv(io.BytesIO(load.content))
election.head()
```

```python
election['Diff']=(election['Obama']-election['Romney'])/100
election.head()
```

#### Problem 3(b)

Make a plot of the differences for the week before the election (e.g. 5 days) where the days are on the x-axis and the differences are on the y-axis.  Add a horizontal line showing 3.9%: the difference between Obama and Romney on election day.

```python
max(election['Start Date'])
```

```python
#your code here
week_before= election[(election['End Date'] > '2012-10-26') & (election['End Date'] <= '2012-10-31')]
plt.figure(figsize=(20,10))
plt.scatter(pd.to_datetime(week_before['End Date'].values),week_before.Diff.values)
plt.axhline(y=0.039,c='r')
plt.xlabel("Days")
plt.ylabel('Differences')
plt.title('the differences for the week before the election')
plt.xticks( rotation='vertical')
plt.show()
```

#### Problem 3(c) 

Make a plot showing the differences by pollster where the pollsters are on the x-axis and the differences on the y-axis. 

```python
#your code here
pollsters =range(len(week_before.Pollster.values))
plt.figure(figsize=(20,10))
plt.scatter(pollsters,week_before.Diff.values)
plt.xticks(pollsters, week_before.Pollster.values)
plt.xlabel("Pollsters")
plt.ylabel('Differences')
plt.title('The differences by pollster ')
plt.xticks( rotation='vertical')
plt.show()
```

Is the *across poll* difference larger than the *between pollster* difference? 

Answer:  between pollster difference is larger

#### Problem 3(d)

Take the average for each pollster and then compute the average of that. Given this difference how confident would you have been of an Obama victory?

**Hint**: Compute an estimate of the SE of this average based exclusively on the observed data. 

```python
#your code here
pollster_mean=np.mean(election.groupby(election.Pollster).mean()['Diff'].values)
pollster_std=np.std(election.groupby(election.Pollster).mean()['Diff'].values)
print('mean:',pollster_mean)
print('std:',pollster_std)
```

Answer: Since Mean is smaller than SE, I have not confident of an Obama victory

#### Problem 3(e)

Show the difference against time and see if you can detect a trend towards the end. Use this trend to see if it improves the final estimate.

```python
#your code here
plt.figure(figsize=(16, 8))
days=election[election['End Date']>='2012-01-01'].groupby('End Date').mean()
plt.plot(pd.to_datetime(days.index).values, days['Diff'].values)
plt.xlabel('Time')
plt.ylabel('Difference')
plt.title('the difference against time')
plt.axhline(y=0,c='r')
plt.show()
```
